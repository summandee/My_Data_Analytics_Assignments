7. Show top 10 employees having (overtimepay+OtherPay) in bar plot?  exclude ('Not Provided', 'Not provided')

```python
import matplotlib.pyplot as plt

# Exclude 'Not Provided' and 'Not provided' in JobTitle
filtered_df = df[~df['JobTitle'].isin(['Not Provided', 'Not provided'])]

# Calculate the sum of OvertimePay and OtherPay
filtered_df['OvertimePlusOtherPay'] = filtered_df['OvertimePay'].astype(float) + filtered_df['OtherPay'].astype(float)

# Get the top 10 employees
top_10_employees = filtered_df.nlargest(10, 'OvertimePlusOtherPay')

# Plot the bar plot
plt.figure(figsize=(12, 6))
plt.bar(top_10_employees['EmployeeName'], top_10_employees['OvertimePlusOtherPay'], color='skyblue')
plt.xlabel('Employee Name')
plt.ylabel('OvertimePay + OtherPay')
plt.title('Top 10 Employees with Highest OvertimePay + OtherPay')
plt.xticks(rotation=45, ha='right')
plt.show()
```

8. Show The Average of BasePay, OverTime, OtherPay for all employees in pie chart? exclude ('Not Provided', 'Not provided')


```python
# Convert BasePay, OvertimePay, and OtherPay to numeric, forcing errors to NaN
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'], errors='coerce')
df['OtherPay'] = pd.to_numeric(df['OtherPay'], errors='coerce')

# Exclude 'Not Provided' and 'Not provided' in JobTitle
filtered_df = df[~df['JobTitle'].isin(['Not Provided', 'Not provided'])]

# Calculate the average of BasePay, OvertimePay, and OtherPay
average_base_pay = filtered_df['BasePay'].mean()
average_overtime_pay = filtered_df['OvertimePay'].mean()
average_other_pay = filtered_df['OtherPay'].mean()

# Data for the pie chart
labels = ['BasePay', 'OvertimePay', 'OtherPay']
sizes = [average_base_pay, average_overtime_pay, average_other_pay]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Average of BasePay, OvertimePay, and OtherPay for All Employees')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

9. Get the average TotalPayBenefits for each JobTitle, but only for those with an average pay greater than 75000? exclude ('Not Provided', 'Not provided')


```python
# Group by JobTitle and calculate the mean of TotalPayBenefits
average_pay_by_jobtitle = df.groupby('JobTitle')['TotalPayBenefits'].mean()

# Filter out JobTitles with an average TotalPayBenefits greater than 75000
filtered_average_pay = average_pay_by_jobtitle[average_pay_by_jobtitle > 75000]

filtered_average_pay
```

10. Get the number of employees (EmployeeName) for each JobTitle, where the count is greater than 10?  exclude ('Not Provided', 'Not provided')


```python
# Exclude 'Not Provided' and 'Not provided' in JobTitle
filtered_df = df[~df['JobTitle'].isin(['Not Provided', 'Not provided'])]

# Group by JobTitle and count the number of employees
employee_count_by_jobtitle = filtered_df.groupby('JobTitle')['EmployeeName'].count()

# Filter JobTitles with a count greater than 10
jobtitles_with_more_than_10_employees = employee_count_by_jobtitle[employee_count_by_jobtitle > 10]

jobtitles_with_more_than_10_employees
```
